# La legge di Rayleigh-Jeans

La legge sulle radiazioni di Rayleigh-Jeans è stata un tentativo utile ma non completamente riuscito di stabilire la forma funzionale degli spettri di radiazione termica. La densità di energia $u_\nu$ per unità di intervallo di frequenza a una frequenza $\nu$ è, secondo The Rayleigh-Jeans Radiation,

$$
u_\nu(T) = \frac{8\pi}{c^2}\nu^2k_bT
$$

* Dove $k$ è la costante di Boltzmann, $T$ è la temperatura assoluta del corpo e $c$è la velocità della luce ne vuoto.
* Nel S.I. l'unità di misura di $u_\nu(T)$ è
$$\frac{J}{m^2},$$ ovvero è un flusso di energia.

Questa formula si adatta ai dati sperimentali per basse frequenze ma fallisce per le frequenze più alte. L'incapacità della formula di abbinare i nuovi dati è stata chiamata *catastrofe ultravioletta*. Il significato di questa inadeguata cosiddetta legge è che fornisce una condizione asintotica che altre formule proposte, come quella di Planck, devono soddisfare. Dà valore a una costante altrimenti arbitraria nella formula della radiazione termica di Planck.

# Derivazione
---

## le ipotesi di Rayleigh-Jeans <a id="iprj"></a> 

### 1. Il campo elettromagnetico in una cavità (corpo nero) all'aequilibrio termico è formato da onde stazionarie con dei nodi al bordo della scatola
$$
\vec E(\vec r)= E_0\sin(\vec k \cdot \vec r)\sin(\omega t)
$$
Con 
$$
\left\{\begin{matrix}
 \vec k =(k_x,k_y,k_z) \\ 
 |\vec k|=\frac{2 \pi}{\lambda}=\frac{\omega}{c}= \frac{2 \pi \nu}{c}
\end{matrix}\right.
;\quad \text{e}\quad \vec r =(r_x,r_y,r_z)
$$

Se abbiamo una scatola di spigolo $L$ allora dovremo avere che
$$
\vec E(\vec L) = 0, \quad \text{dove}\quad \vec L = (L,L,L).
$$

### 2. Vale il teorema di equipartizione e quindi ogni oscillatore armonico possiede un'energia media 
$$<\epsilon> = \frac{1}{2}kT$$

---

## Numero di modi oscillatori stazionari <a id="rj.1"></a>

### Onde stazionarie nella cavita
Ora bisogna ricavare il nuemro di oscillatori in una data lunghezza d'onda applicando la condizione di stazionarietà e considerando come cavità una scatola cubica di spigolo $L$.

Avremo onde stazionarie 
$$
\sin(k_x L)=\sin(k_y L)=\sin(k_z L)=0
$$

Quindi
$$
\left\{\begin{matrix}
 k_x L = n_x \pi  \\ 
 k_y L = n_y \pi  \\
 k_z L = n_z \pi
 \end{matrix}\right.
\Rightarrow
\left\{\begin{matrix}
 k_x = n_x \frac{\pi}{L}  \\ 
 k_y = n_y \frac{\pi}{L} \\
 k_z = n_z \frac{\pi}{L}
 \end{matrix}\right. \quad \vec n=(n_x,n_y,n_z)\in \mathbb{N}\; \ge 0 \quad n = \sqrt{(n_x+n_y+n_z}
$$
Da cui otteniamo

$$
k^2 = \pi^2 \left ( \left(\frac{n_x}{L}\right )^2+\left(\frac{n_y}{L} \right )^2+\left(\frac{n_z}{L} \right )^2 \right )
$$
Che si riduce a
$$
k^2 = \frac{\pi^2n^2}{L^2}\\
\left (\frac{2 \pi}{\lambda} \right )^2 = \frac{\pi^2n^2}{L^2} \\
\frac{4 \pi^2}{\lambda^2} = \frac{\pi^2n^2}{L^2}\\
\frac{4 }{\lambda^2} = \frac{n^2}{L} \Rightarrow n^2 = \frac{4 L^2}{\lambda^2} = 4 L^2\frac{\nu^2}{c^2}
$$

---

### numero di modi

Dobbiamo valutare il numero di modi che possono soddisfare questa condizione, il che equivale a contare tutte le possibili combinazioni dei valori interi $n$. Un'approssimazione può essere fatta trattando il numero di combinazioni come il volume di una griglia tridimensionale dei valori di $n$, di un "$n$-spazio" 3-dimensionale. Usando la relazione per il volume di una sfera, con i valori $n$ che specificano le coordinate lungo tre assi $n$, si ottiene
<img width="400" alt="The Rayleigh scheme for counting modes.
After Richtmyer, et al." src="http://hyperphysics.phy-astr.gsu.edu/hbase/quantum/imgqua/nspace.png">
Il numero di modi che cerchiamo $dN$ che si trova nell'intervallo $(n,n+dn)$. Cerchiamo quindi il volume di modi in una calotta sferica nello spazio dei modi $n$. Cioè:

$$
dN = 4 \pi n^2 dn
$$

Ricavando
$$
dn = \frac{2L}{\lambda^2} d\lambda = \frac{2L}{c}d\nu \quad \text{in modulo}\\
$$

Che sostiuendo ci da:

$$
\left\{\begin{matrix}
 dN = 4 \pi \frac{4 L^2}{\lambda^2}  \frac{2L}{\lambda^2} d\lambda =32\pi  \frac{ L^3 }{\lambda^4} d\lambda\\ 
dN = 4 \pi \frac{4 L^2}{c^2}\nu^2 \frac{2L}{c}d\nu = 32\pi\frac{L^3}{c^3}\nu^2 d\nu
 \end{matrix}\right.
$$


Ma ricordando che $n_x,n_yn_z \ge 0$ allora dobbiamo prendere solo un ottante della sfera per cui:
$$
\left\{\begin{matrix}
 dN  =4\pi  \frac{ L^3 }{\lambda^4} d\lambda\\ 
dN = 4 \pi\frac{L^3}{c^3}\nu^2 d\nu
 \end{matrix}\right.
$$

### energia media

L'energia cinetica media per grado di libertà è $\frac{1}{2}kT$, dove $k$ è la costante di Boltzmann. Per gli oscillatori armonici esiste un'uguaglianza tra energia cinetica e potenziale, quindi l'energia media per grado di libertà è $kT$. 

$$
<\epsilon_n> = kT
$$
Considerando che gli oscillatori sul bordo della cavità possono oscillare in sole due direzioni (due direzioni di polarizzazione) , hannno quindi due gradi di libertà, allora
$$
<\epsilon_n> = 2kT
$$


Ciò significa che l'energia media di radiazione $E$ è data da

$$
<dE> = \text{numero di oscillatori}\times\text{energia media degli oscillatori}\\
<dE> = dN<\epsilon_n> 
$$

Cioè
$$
<dE> = 8\pi  \frac{ L^3 }{\lambda^4} d\lambda kT\\
\quad \\
<dE> = 8\pi\frac{L^3}{c^3}\nu^2 d\nu kT
$$
L'energia per unità di lunghezza d'onda o di frequenza è data rispettivamente da:
$$
\frac{dE}{d\lambda} =  8\pi  \frac{ L^3 }{\lambda^4} kT\\
\quad \\
\frac{dE}{d\nu} =  8 \pi\frac{L^3}{c^3}\nu^2 kT
$$

Per ottenere la densità di energia, ovvero energia per unità di volume, basta dividere per il volume $V=l^3$ è quindi data da

$$
u_\lambda(T) =  \frac{ 8\pi }{\lambda^4} kT \quad \text{in termini di }\lambda;\\
\quad \\
\quad \\
u_\nu(T) =  \frac{8 \pi}{c^3}\nu^2 kT \quad \text{in termini di }\nu.
$$




# Il grafico


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import h, c, k, pi
# the following must be imported for interactive plots
from ipywidgets import interact, interactive, fixed
```

```python
def R_J_spectral_density(frequency, temperature):
    """Spectral energy density according to Rayleigh-Jeans's law.
    Uses the SI unit system, so the frequency must be in meters (m) and the
    temperature must be in Kelvin (K). Returns the spectral energy density form
    of the Rayleigh-Jeans's law in Joule per cubic meter per spectral unit.

    Parameters
    ----------
   frequency : float or array
        The frequency(Hz) in Hz. Accepts vectors or floats
    temperature : float
        Temperature in Kelvin.

    Returns
    -------
    array or float
        Spectral energy density vector or float (depends on frequency)
    """
    #define Rayleigh-Jeans fnction
    rho = (8*pi/c**2)*frequency**2*k*temperature
    
    return rho
```

```python
def plot_visible(lines=100, transparency=0.3, linewidth=3, unit_exponent=1e9):
    """Plots a visible spectrum in the current axis. Must be called before the
    desired plot.

    Parameters
    ----------
    lines : int, optional
        The number of lines. Increase if blank spaces are noted, by default 100
    transparency : float, optional
        The transparency of the lines, by default 0.3
    linewidth : int, optional
        The width of lines, by default 3. Be careful not to use a large number
    unit_exponent : float, optional
        Usually the plots are in nanometers so by default 1e9.
    """
    ax = plt.gca()
    steps = lines
    visible = np.linspace(4e14, 8e14, steps)
    colormap = plt.cm.gist_rainbow
    colors = [colormap(i) for i in np.linspace(0.0, 1.0, steps)]
    j = 0
    for val in visible:
        ax.axvline(val * unit_exponent, lw=linewidth,
                   color=colors[-j], alpha=transparency, zorder=-1)
        j += 1
```

```python
def plot_RJ(frequency_array,
                temperature_array,
                colors=plt.cm.coolwarm,
                tick_fontsize=14,
                axes_fontsize=14,
                title_fontsize=16):
    """Plots the Rayleigh-Jeans's law. The plot is in SI units (frequency will appear
    in nanometer).

    Parameters
    ----------
    frequency_array : array
        frequency array in meters
    temperature_array : array
        Temperature in Kelvin. If only one, pass as a list with the value,
        e.g. [5000]
    colors : [type], optional
        Desired colormap, by default plt.cm.coolwarm
    tick_fontsize : int, optional
        Tick labels font size, by default 14
    axes_fontsize : int, optional
        Axes labels font size, by default 14
    title_fontsize : int, optional
        Title font size, by default 16
    """

    results = []
    for temperature in temperature_array:
        results.append(R_J_spectral_density(frequency_array, temperature))

    # gets the current axes and set a colormap
    ax = plt.gca()

    colormap = colors
    ax.set_prop_cycle(plt.cycler('color', colormap(
        np.linspace(0, 1, len(temperature_array)))))

    # grid lines
    ax.grid(linestyle=':', linewidth=1.5)

    # the plots and legend
    for result, temperature in zip(results, temperature_array):
        ax.plot(frequency_array * 1e12, result,
                label='{} K'.format(temperature), linewidth=3)
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=14)

    # setting the y-axis to scientific notation and
    # getting the order of magnitude
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
    ax.yaxis.major.formatter._useMathText = True
    ax.figure.canvas.draw()  # Update the text
    order_magnitude = ax.yaxis.get_offset_text().get_text().replace('\\times',
                                                                    '')
    ax.yaxis.offsetText.set_visible(False)

    # labels and title
    ax.set_xlabel('Frequency / Hz', fontsize=axes_fontsize)
    ax.set_ylabel('Energy density / (' + order_magnitude +
                  ' $J/m^3$)', fontsize=axes_fontsize)
    ax.tick_params(labelsize=tick_fontsize)
    ax.set_title('Rayleigh-Jeans law - black body radiation',
                 fontsize=title_fontsize)

    plt.tight_layout()
    plt.show()

```

```python
R_J_spectral_density(300.0e-9, 1000)
```

```python
frequency_array = np.linspace(3, 3e6, 1000)  
temperature_array = np.arange(1000, 7001, 500)


fig1 = plt.figure(figsize=(10, 6))
ax = fig1.add_subplot(111)
plot_RJ(frequency_array , temperature_array)
plt.savefig('Rayleigh-Jeans.png')
plt.savefig('Rayleigh-Jeans.pdf')
plt.savefig('Rayleigh-Jeans.svg')
```
